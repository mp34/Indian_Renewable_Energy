# Heat Map Biomass Reneable Energy statewise year(2020)

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/mihir/Desktop/BIDA/Tableau/Indian_renewable_energy/Python/Biomass.csv")
df.head(10)
```

```python
States = ((np.array(df['State'])).reshape(4,7))
Capacities= ((np.array(df['Capacity'])).reshape(4,7))
print(States)
print(Capacities)
```

```python
result = df.pivot(index  = 'yrows',columns= 'xcols',values = 'Capacity')
print(result)
```

```python
labels = (np.array(["{0}\n {1:.2f}".format(state,capacities)
                   for state, capacities in zip(States.flatten(),Capacities.flatten())]).reshape(4,7))
```

```python
fig,ax = plt.subplots(figsize = (20,20))


title = "Biomass Energy Capacity"
plt.title(title,fontsize=18)
tt1 = ax.title
tt1.set_position([0.5,1.05])
ax.set_xticks([])
ax.set_yticks([])
ax.axis("off")

sns.heatmap(result,annot=labels,fmt="",cmap="Greens",linewidths=2,ax=ax,linecolor='green')
plt.show()
```
